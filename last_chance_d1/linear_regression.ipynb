```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### California_Housing sckit-learn dataset

```python
from sklearn.datasets import fetch_california_housing
```

```python
california_data = fetch_california_housing()
print(california_data)
```

```python
print(california_data.feature_names)
```

#### Feature Columns: Input to the model ( x varibles )

```python
dataset = pd.DataFrame(california_data.data, columns=california_data.feature_names)
dataset.head()
```

```python
dataset.shape
```

#### Target Column: Single Column for the output( y variable )

```python
# Added "Price" as target column in the dataset 
dataset['Price'] = california_data.target
dataset.head
```

**Check if there is any null value in the dataset** 

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

**Split dataset into train and test: x is input or feature column and y is output or target column**

```python
x = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train
```

```python
x_test
```

**Now we will use linear regression model**

```python
from sklearn.linear_model import LinearRegression
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)
```

**Check Accuracy**

```python
regression_model.score(x_test, y_test)
```

```python
regression_model.coef_    ##Checks the value of b0, b1, b2......(Linear Equation)
```

```python
regression_model.intercept_    ##Checks the value of constant from Linear Equation
```

#### Standardiaztion the dataset(Convert data in between 0 - 1)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
##Scaling the x_train and x_test data
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

**Fitting the scaled data into the model to check if the accuracy improved or not**

```python
regression_model.fit(x_train_scaled, y_train)
```

**Predict the model using x_test**

```python
predict_regression = regression_model.predict(x_test_scaled)
```

```python
##Now check the score again
regression_model.score(x_test_scaled, y_test)
```

**Check Mean Squared Error and Mean Absolute Error**

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

MSE and MAE near zero(0) means better accuracy

```python
print("MSE:", mean_squared_error(y_test, predict_regression))
print("MAE:", mean_absolute_error(y_test, predict_regression))
##In this case the accuracy is bad as hell
```

```python
##RMSE: Root mean squared error
print("RMSE:", np.sqrt(mean_squared_error(y_test, predict_regression)))
```
